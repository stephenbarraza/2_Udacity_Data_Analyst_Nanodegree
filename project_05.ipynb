### What Variables Impact the Interest Rate & Annual Percentage Rate (APR) of Prosper's Loans?<br>

The project will focus on generally accepted, industry-wide practices around interest rates & APR determinations in the loan servicing industry using Prosper Loans as a specifc data point in that industry.

#### What is an Interest Rate & an APR?<br>

Per [Better.com](https://better.com/content/difference-between-mortgage-apr-and-interest-rate/?utm_source=google&utm_medium=search&utm_campaign=Search_G_NonBrand_DSA_App&utm_content=agn:DSA_App_All___m:b_d:c_n:g_tid:dsa-130412282889&utm_term=&gclid=Cj0KCQiAyJOBBhDCARIsAJG2h5fLhgfKkcqX4n1zyjWGuZ2m1Hpo2e4EuqNbJHmEISnZ6IRhkfCxVR8aAl0FEALw_wcB), interest rate & APR both describe the cost of borrowing money.  A loan’s interest rate is expressed as a percentage of the principal loan amount that a lender charges you on an annual basis for borrowing your principal loan amount.  The interest rate a borrower can "lock in" is based on a number of factors, some outside the borrower's control (inflation, economic growth, the bond market, and housing conditions) and other's within their control (the borrower's credit score, down payment, loan amount, and debt-to-income ratio).  A loan’s annual percentage rate (APR) is also displayed as a percentage, but is higher than the accompanying interest rate. That’s because an APR not only takes your interest rate into account, but also factors in other costs, such as most closing costs and lender fees. 

#### What levels of play do these variables exhibit on Prosper Loan business?

With Prosper Loan data in hand, the focus will be to determine what exact, in what combination, and at what level these factors come to determine the Interest Rate and/or APR that Prosper charges their customers.

<a id='measurement_levels'></a>
### What are Measurement Levels and What's the Need for Pre-Analysis?

Levels of measurement are ways of categorizing and boxing up data for easier understanding and analysis.  These measurement levels (explained in [course work](https://classroom.udacity.com/nanodegrees/nd002/parts/9f7e8991-8bfb-4103-8307-3b6f93f0ecc7/modules/bc709f85-0ebb-45b8-907b-065adc25cbce/lessons/a755a07d-345f-4e57-91c2-3b3b9ae7cc28/concepts/2d20613b-c4a6-452e-8a70-ca24b10294a) & [elsewhere](https://www.questionpro.com/blog/nominal-ordinal-interval-ratio/)) can be summed up as follows: 

- Qualitative Nominal Level: a naming scale.<br>


- Qualitative Ordinal Level: a naming scale and a specific ordering scale.<br>


- Quantitative Interval Level: a naming scale, a specific ordering scale and a specific interval between each category of the varaible. Addition & subtraction can be performed and absolute differences are meaningful (an +/- of 1 pt will mean the same regardless of where you are on the scale; it can accomodate negative numbers).<br>


- Quantitatitve Ratio Level:  a naming scale, a specific ordering scale and a specific interval between each category of the variable. Addition, subtraction, mutiplication & division can be performed and relative differences are meaningful. This level can also accommodate the value of “zero” on any of its variables.  Has no option for negative numbers.<br>

Also, if varaiables are in either of the Quantitative Levels then they are also either:

- Discrete Varaibles: take on a set value at a maximum level of precision (either they are or aren't at the set value); values can't be subdivided.<br>


- Continuous Variables: take on a value at any level of precision; values can be subdivided to infinity.<br>


By understanding these levels, we can "pre-analyze" and box up all the variables in this study according to their measurement level.  By doing this first, we'll better understand what set of variables should be compared to other varaiables during actual Exploratory Analysis phase later on.
